```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.svm import SVR
```

```python
mesh_size = 0.2
margin =0
df = px.data.iris()
X = df[['sepal_width', 'sepal_length']]
y = df['petal_width']
model = SVR(C=1.0)
model.fit(X,y)
```

```python
x_min, x_max=X.sepal_width.min()-margin,X.sepal_width.max()+margin
y_min, y_max=X.sepal_length.min()-margin,X.sepal_length.max()+margin
xrange = np.arange(x_min, x_max, mesh_size)
yrange = np.arange(y_min, y_max, mesh_size)
xx,yy=np.meshgrid(xrange,yrange)
```

```python
pred = model.predict(np.c_[xx.ravel(), yy.ravel()])
pred = pred.reshape(xx.shape)
```

```python
fig = px.scatter_3d(df,x='sepal_width', y='sepal_length', z = 'petal_width')
fig.update_traces(marker=dict(size=5))
fig.add_traces(go.Surface(x=xrange, y=yrange, z=pred, name='pred_surface'))
fig.show()
```

```python
import plotly.subplots as sp
import plotly.graph_objects as go
import pandas as pd
```

```python
# Dados de vendas
vendas = [5, 10, 14, 20, 25, 30, 35, 50, 100, 40, 150]
meses = ['Janeiro', 'Fevereiro', 'Maro', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro']
```

```python
# Mapear os nomes dos meses para nmeros de ms
meses_numeros = {'Janeiro': 1, 'Fevereiro': 2, 'Maro': 3, 'Abril': 4, 'Maio': 5, 'Junho': 6, 'Julho': 7, 'Agosto': 8, 'Setembro': 9, 'Outubro': 10, 'Novembro': 11}
```

```python
# Criar dataframe para os dados
df = pd.DataFrame({'Ms': meses, 'Vendas': vendas})
# Criar subplots
fig = sp.make_subplots(rows=2, cols=2,
                       subplot_titles=('Vendas Mensais', ' Vendas Mensais'),
                       specs=[[{'type': 'bar'}, {'type': 'indicator'}],
                              [{'type': 'table', 'colspan': 2}, None]])
```

```python
# Grfico de barras
fig.add_trace(go.Bar(x=df['Ms'], y=df['Vendas']), row=1, col=1)
fig.update_xaxes(title_text='Ms', row=1, col=1)
fig.update_yaxes(title_text='Vendas', row=1, col=1)
# Grfico de gauge
fig.add_trace(go.Indicator(
    mode='gauge',
    value=df['Vendas'][10],
    domain={'x': [0, 0.4], 'y': [0.4, 0.5]},
    title={'text': ""},
    gauge={'axis': {'range': [None, max(vendas)]},
           'bar': {'color': 'darkblue'},
           'steps': [
               {'range': [0, 50], 'color': 'lightgray'},
               {'range': [50, 100], 'color': 'gray'}],
           'threshold': {'line': {'color': 'red', 'width': 4}, 'thickness': 0.75, 'value': max(vendas)}}), row=1, col=2)
```

```python
# Tabela de vendas mensais
df_table = df.copy()
df_table['Ms'] = df_table['Ms'].map(meses_numeros)
df_table = df_table.sort_values('Ms')
df_table = df_table.reset_index(drop=True)
fig.add_trace(go.Table(
    header=dict(values=['Ms', 'Vendas'],
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[df_table['Ms'], df_table['Vendas']],
               fill_color='lavender',
               align='left')), row=2, col=1)

fig.update_layout(height=800, width=800, showlegend=False,
                  plot_bgcolor='white',
                  paper_bgcolor='white',
                  font=dict(color='black'))
fig.write_html('Dashteste1')
fig.show()
```
